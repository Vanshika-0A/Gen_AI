Window Based Model

GloVe - Window Based Model (Context Usage)

After GloVe we have:
word -> vector
But NLP tasks won't be for oe single word.
They are about word + surrounding words
So we need to use a Window

Window = fixed number of words around a target word

```python
import gensim.downloader as api
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
print("Loading the model...")
model = api.load("glove-wiki-gigaword-50")
print("Model loaded successfully!")
```

```python
sentence = "Please open google and check weather"
tokens = word_tokenize(sentence.lower())
```

```python
#sliding window function
def get_window(tokens, window_size=2):
    windows = []
    for i in range(len(tokens)):
        left = max(0, i - window_size)
        right = min(len(tokens), i + window_size + 1)
        window_words = tokens[left:right]
        windows.append(window_words)
    return windows
       
```

```python
windows = get_window(tokens, 2)
print("Window\n,", windows)
```

```python
def window_vector(window):
    vectors = []
    for word in window:
        if word in model:
            vectors.append(model[word])
    if len(vectors)==0:
        return np.zeroes(100)
    else:
        return np.mean(vectors)
```

```python
window_vectors = [window_vector(w) for w in windows]
print(window_vectors[0].shape)
```
