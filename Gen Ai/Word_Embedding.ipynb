```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
sentence = ["I am leaning natural language processing",
            "Natural language processing is a part of machine learning",
            "I enjoy machine learning",
            "Deep learning uses nueral networks",
            "Neural networks is powerful for natural language procsessing"
             ]
```

```python
#Tokenize 
tokenized_sentences = [word_tokenize(s.lower()) for s in sentence ]
print(tokenized_sentences)
```

```python
model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1)
```

```python
print(model.wv.similarity('natural','language'))
```

```python
print(model.wv.most_similar('learning'))
print(model.wv.most_similar('machine',topn=5))
```
